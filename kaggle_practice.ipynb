```python
import kagglehub 

#Downloads the most recent dataset from Kaggle. 
path = kagglehub.dataset_download("dhrubangtalukdar/fortune-500-companies-stock-data")

print("Path to dataset files:", path) #Just a string of where Kaggle downloads to.
```

```python
import os 

source = os.listdir(path+ "/fortune_500_stock_data")
```

```python
source
```

```python
import pandas as pd 

df_dict = {}

for file in source:
    df_dict[file] = pd.DataFrame(pd.read_csv(path+"/fortune_500_stock_data/"+file))
```

```python
ibm = df_dict['IBM.csv']
ibm
```

```python
#Get rid of Garabge at the top of the df 
ibm = ibm.drop(index=0).reset_index(drop=True)
```

```python
ibm
```

```python
for col in ibm.columns:
    try: 
        ibm[col] = pd.to_numeric(ibm[col])
    except: pass
```

```python
ibm.dtypes
```

```python
ibm
```

```python
ibm.plot(x="Date",y=["High","Low"])
```

```python
ibm.query("Close > 200 and Close < 350")
```

```python
ibm["Diff"] = ibm["Open"] - ibm["Close"]
```

```python
ibm["Diff"] = abs(ibm["Diff"])
```
